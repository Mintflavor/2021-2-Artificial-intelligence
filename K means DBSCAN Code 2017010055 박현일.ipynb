```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
%matplotlib inline
cmap = "tab10"
```

```python
X = -3 * np.random.rand(150, 3)
X1 = 3 + 3 * np.random.rand(50, 3)
X2 = 3 * np.random.rand(50, 3)
X[50:100, :] = X1
X[100:150, :] = X2
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

# #1 K-means

```python
K_means = KMeans(n_clusters=3, random_state=0)
K_means.fit(X)

Y_pred = K_means.fit_predict(X)
centers = K_means.cluster_centers_
print(Y_pred)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y_pred, cmap=cmap)
plt.scatter(centers[:, 0], centers[:, 1], s=100, c='r', marker='s')
plt.show()
```

# #2 DBSCAN

```python
enu = ((0.5, 3), (0.5, 5), (1, 3), (1, 5), (1.5, 3), (1.5, 5))
for i, (eps, min_samples) in enumerate(enu):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    Y_pred = dbscan.fit_predict(X)
    plt.figure(i+1)
    plt.scatter(X[:, 0], X[:, 1], cmap=cmap, c=Y_pred)
    plt.title(f"eps = {eps}, min_samples = {min_samples}")
    plt.show()
```
